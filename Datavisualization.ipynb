```python
EV_data["Charging_Start_Time"] = pd.to_datetime(EV_data["Charging_Start_Time"], errors='coerce')
EV_data["Charging_End_Time"] = pd.to_datetime(EV_data["Charging_End_Time"], errors='coerce')
```

```python
invalid_dates = EV_data[EV_data["Charging_Start_Time"].isna()]
print(invalid_dates)
```

```python
EV_data.dropna(subset=["Charging_Start_Time"], inplace=True)
EV_data.dropna(subset=["Charging_End_Time"], inplace=True)
```

```python
EV_data[EV_data["Charging_End_Time"].isna()]
```

```python
EV_data["Hour"] = EV_data["Charging_Start_Time"].dt.hour
EV_data["Month"] = EV_data["Charging_Start_Time"].dt.month
EV_data["Year"] = EV_data["Charging_Start_Time"].dt.year
```

```python
EV_data["Charging_Duration"] = (

    EV_data["Charging_End_Time"] - EV_data["Charging_Start_Time"]

).dt.total_seconds() / 3600
```

```python
EV_data['Charging_Duration_(hours)'] = (EV_data['Charging_End_Time'] - EV_data['Charging_Start_Time']).dt.total_seconds() / 3600
```

```python
EV_data['Charging_Rate_(kW)'] = EV_data['Energy_Consumed_(kWh)'] / EV_data['Charging_Duration_(hours)']
```

```python
plt.figure(figsize=(10, 6))
plt.hist(EV_data['Energy_Consumed_(kWh)'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Distribution of Energy Consumed')
plt.xlabel('Energy Consumed (kWh)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(EV_data['Battery_Capacity_(kWh)'], EV_data['Energy_Consumed_(kWh)'], alpha=0.7)
plt.title('Battery Capacity vs. Energy Consumed')
plt.xlabel('Battery Capacity (kWh)')
plt.ylabel('Energy Consumed (kWh)')
plt.show()
```

```python
# Group by 'Model' and calculate mean energy consumed
avg_energy_per_model = EV_data.groupby('Model')['Energy_Consumed_(kWh)'].mean()

# Plot bar chart
plt.figure(figsize=(10, 6))
avg_energy_per_model.plot(kind='bar', color='skyblue')
plt.title('Average Energy Consumed per Vehicle Model')
plt.xlabel('Vehicle Model')
plt.ylabel('Average Energy Consumed (kWh)')
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding values on top of bars
for index, value in enumerate(avg_energy_per_model):
    plt.text(index, value, f'{value:.2f}', ha='center', va='bottom', fontsize=10, color='blue')

plt.show()
```

```python
avg_cost_per_day = EV_data.groupby('Day_of_Week')['Charging_Cost_(USD)'].mean()

# Plot line chart
plt.figure(figsize=(10, 6))
avg_cost_per_day.plot(kind='line', marker='o', color='green')
plt.title('Average Charging Cost over Days of the Week')
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Average Charging Cost (USD)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

# Adding values on data points
for i, value in enumerate(avg_cost_per_day):
    plt.text(i, value, f'{value:.2f}', ha='center', va='bottom', fontsize=10, color='red')

plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(EV_data['Battery_Capacity_(kWh)'], EV_data['Energy_Consumed_(kWh)'], alpha=0.7, color='purple')
plt.title('Battery Capacity vs. Energy Consumed')
plt.xlabel('Battery Capacity (kWh)')
plt.ylabel('Energy Consumed (kWh)')
plt.grid(True)
plt.show()
```

```python
# Count user types
user_type_counts = EV_data['User_Type'].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 8))
user_type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Proportion of Different User Types', fontsize=16)
plt.ylabel('')  # Removing the y-label

# Adding a legend
plt.legend(user_type_counts.index, title='User Type', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
plt.hist(EV_data['Charging_Duration_(hours)'], bins=30, edgecolor='black', alpha=0.7, color='orange')
plt.title('Distribution of Charging Duration')
plt.xlabel('Charging Duration (hours)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(10, 6))
plt.boxplot(EV_data['Charging_Cost_(USD)'])
plt.title('Box Plot of Charging Costs')
plt.ylabel('Charging Cost (USD)')
plt.grid(True)
plt.show()
```
